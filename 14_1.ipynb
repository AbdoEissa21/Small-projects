```python
def check_mat(a):
    if a.shape[0]== a.shape[1]:
        return True
    else:
      return False

def stat(x,axis='col'):
  mat=np.array(x)
  check_mat(mat)
  if(axis=='col'):
    mean=np.mean(mat)
    variance=np.var(mat)
    standard=np.std(mat)
  elif(axis=='raw'):
      mean=np.mean(mat,axis='raw')
      variance=np.var(mat='raw')
      standard=np.std(mat='raw')
  else:
    return 'Error'
  return mean,variance,standard
```
