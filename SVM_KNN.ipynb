```python
import numpy as np
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
```

```python
df=pd.read_csv('train (1).csv')
```

```python
df.head
```

```python
df.isnull().sum()
```

```python
data=df.dropna()
```

```python
data.isnull().sum()
```

```python
data.describe()
```

```python
data.info()
```

```python
data.drop(['Name', "Cabin", "Embarked", 'Ticket','Sex'], axis=1, inplace= True)
```

```python
data.head()
```

```python
X = data.iloc[:, 2:]
```

```python
X
```

```python
y = data['Survived']
# y=y.iloc[:183]
```

```python
X.shape,y.shape
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20,
                                                    random_state=1,
                                                    shuffle=True)
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
```

```python
knn.fit(X_train, y_train)
```

```python
# knn.score(X_train, y_train) #with knn =2,4,5 the best knn is 3 
```

```python
knn.score(X_train, y_train)
```

```python
knn.predict(X_test)
```

```python
knn.fit(X_test, y_test)
```

```python
knn.score(X_test, y_test)
```
